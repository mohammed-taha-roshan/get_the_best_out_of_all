```python
# for ed,employ,equip,call card,qireless
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
# %matplotlib inline 
import matplotlib.pyplot as plt
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/ChurnData.csv")

churn_df = df[["ed","employ","equip","callcard","wireless"]]


X = np.asanyarray(churn_df[["ed","employ","equip","callcard","wireless"]])


y = np.asanyarray(df['churn'])


from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR

yhat = LR.predict(X_test)
yhat

yhat_prob = LR.predict_proba(X_test)
yhat_prob

arr=confusion_matrix(y_test, yhat)
# accuracy=
print(arr)
print((arr[0][0]+arr[1][1])/(arr[0][0]+arr[1][1]+arr[0][1]+arr[1][0]))
```
